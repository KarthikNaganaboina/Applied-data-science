```python
import pandas as pd
import wbgapi as wb
import matplotlib.pyplot as plt
```

```python
#Function using pandas to load the dataset values from the World Indicator dataset
datafrm=pd.read_csv(r"C:\\Kartik ADS2\World_Bank_Data.csv", low_memory=False)
```

```python
#Initial rows of the dataset
datafrm.head(7)
```

```python
#Performing the index setting
datafrm1=datafrm.set_index('economy')
```

```python
#Transposed data after setting the index
datafrm1.T.head(7)
```

```python
#Economic and climate indicators for the analysis
Econ_IND = ['NY.GDP.MKTP.CD','SP.POP.TOTL','SL.UEM.1524.FE.ZS','NE.IMP.GNFS.ZS']
country_code = ['BRA',"GBR",'IND','BGD','AUS','FRA','CHE','LUX','JPN','JAM']
Clim_IND=['EG.ELC.NGAS.ZS','EG.ELC.RNWX.KH','EN.ATM.CO2E.PC','EN.ATM.GHGT.KT.CE','EN.ATM.CO2E.PC']
ECONMY  = wb.data.DataFrame(Econ_IND, country_code, mrv=7)
CLMATE = wb.data.DataFrame(Clim_IND, country_code, mrv=7)
#NY.GDP.MKTP.CD: Current GDP of a nation
#SP.POP.TOTL: Total population of a nation
#SL.UEM.1524.FE.ZS: Female unemployment of youths
#NE.IMP.GNFS.ZS: Total import of a nation
#EG.ELC.NGAS.ZS: Electricity production using natural gas
#EG.ELC.RNWX.KH: Electricity production using renewable sources
#EN.ATM.GHGT.KT.CE: Greenshouse gas emissions
#EN.ATM.CO2E.PC: Carbon dioxide emissions in metric tons per capita
```

```python
# Economic indicator analysis
ECONMY.columns = [a.replace('YR','') for a in ECONMY.columns]      
ECONMY=ECONMY.stack().unstack(level=1)                             
ECONMY.index.names = ['Nation_Code', 'Year']                           
ECONMY.columns                                                     
ECONMY.fillna(0)
ECONMY.head(7)
```

```python
# Climate indicator analysis
CLMATE.columns = [a.replace('YR','') for a in CLMATE.columns]      
CLMATE=CLMATE.stack().unstack(level=1)                             
CLMATE.index.names = ['Nation_Code', 'Year']                           
CLMATE.columns                                                     
CLMATE.fillna(0)
CLMATE.head(7)
```

```python
#Eliminating the null values from dataset and resetting proper index
data1=ECONMY.reset_index()
data2=CLMATE.reset_index()
data3=data1.fillna(0)
data4=data2.fillna(0)
```

```python
#Creating the dataframe for analysis by joining
final_data = pd.merge(data3, data4)
final_data.head(7)
```

```python
# Statistical analysis of Brazil
d1=final_data[(final_data['Nation_Code']=='BRA')]
d1.describe()
```

**The average total import in Brazil is equal to 13.22**

```python
# Line plot chart- Brazil's Current GDP
plt.plot(d1["Year"], d1["NY.GDP.MKTP.CD"],color="red")
plt.ylabel("NY.GDP.MKTP.CD")
plt.xlabel("Year")
plt.show()
```
